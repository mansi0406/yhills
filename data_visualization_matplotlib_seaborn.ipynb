```python
import plotly.express as  px
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
titanic_df=pd.read_csv("titanic-training-data.csv")
```

```python
titanic_df
```

```python
#set the figure size
plt.figure(figsize=(8,6))

#create the countplot of passengers by class
sns.countplot(x='Pclass',hue='Sex',data=titanic_df)

#set the title and axis lables
plt.title('Passengers by Class and Sex')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
#set the figure size
plt.figure(figsize=(8,6))

#create the countplot of passengers by class
sns.countplot(x='Embarked',data=titanic_df)

#set the title and axis lables
plt.title('Passengers by embarked')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
#set the figure size
plt.figure(figsize=(8,6))

#create the countplot of passengers by class
sns.countplot(x='Pclass',hue='Embarked',data=titanic_df)

#set the title and axis lables
plt.title('Passengers by pclass and embarked')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
#set the figure size
plt.figure(figsize=(8,6))

#create the violinplot of passengers by class
sns.violinplot(x='Pclass',y='Age',data=titanic_df)

#set the title and axis lables
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
#set the figure size
plt.figure(figsize=(8,6))

#create the boxplot of passengers by class
sns.boxplot(x='Pclass',y='Age',data=titanic_df)

#set the title and axis lables
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
#set the figure size
plt.figure(figsize=(8,6))

#create the swarmplot of passengers by class
sns.swarmplot(x='Pclass',y='Age',data=titanic_df)

#set the title and axis lables
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
titanic_df.hist(figsize=(10,10),color='orange')
plt.show()   #without description
```

```python
corr=titanic_df.corr()
plt.figure(figsize=(6,6))
sns.heatmap(corr,annot=True)     #annot is for text under the box
plt.title("Correlation Matrix")
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```
